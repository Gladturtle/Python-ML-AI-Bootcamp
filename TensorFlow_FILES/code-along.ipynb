```python
from keras.api.models import Sequential
from keras.api.layers import Dense
```

```python
model.fit(x = X_train,y=y_train,validation_data=(X_test,y_test),batch_size=256,epochs=400)
```

```python
loss = pd.DataFrame(model.history.history)
loss.plot()
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,explained_variance_score
```

```python
sns.displot(data= df, x = 'benign_0__mal_1',y = 'mean radius')
```

```python
X = df.drop('benign_0__mal_1',axis = 1)
y = df['benign_0__mal_1']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from keras.api.models import Sequential
from keras.api.layers import Dense
```

```python
model = Sequential()

model.add(Dense(31,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='adam',loss = 'binary_crossentropy')
```

```python
model.fit(x = X_train,y=y_train,validation_data=(X_test,y_test),epochs=600)
```

```python
loss = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=(12,12))
sns.lineplot(loss)
```

```python
from keras.api.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=30)
```

```python
model.fit(x = X_train,y=y_train,validation_data=(X_test,y_test),epochs=600,callbacks=[early_stop])
```

```python
loss = pd.DataFrame(model.history.history)
plt.figure(figsize=(12,12))
sns.lineplot(loss)
```

```python
from keras.api.layers import Dropout
model1 = Sequential()
model1.add(Dense(31,activation='relu'))
model1.add(Dropout(0.5))
model1.add(Dense(15,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
```

```python
model1.compile(optimizer='adam',loss = 'binary_crossentropy')
```

```python
model1.fit(X_train,y_train,batch_size= 64,epochs= 600,callbacks=[early_stop],validation_data = (X_test,y_test))
```

```python
loss = pd.DataFrame(model1.history.history)
plt.figure(figsize=(12,12))
sns.lineplot(loss)
```

```python
pred1 = model1.predict(X_test)
pred = model.predict(X_test)
```

```python
pred = np.where(pred>0.5,1,0)
pred1 = np.where(pred1>0.5,1,0)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
print('\n')
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred1))
print('\n')
print(confusion_matrix(y_test,pred1))
```
