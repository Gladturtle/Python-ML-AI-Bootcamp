```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px


```

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df)
```

```python
sdf = scaler.transform(df)
```

```python
sdf = pd.DataFrame(sdf,columns=df.columns)
```

```python
sdf
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(sdf)
```

```python
pcad = pca.transform(sdf)
pcad = pd.DataFrame(pcad,columns=['1','2','3'])
pcad
```

```python

    


px.scatter_3d(pcad,x='1',y='2',z='3',color=cancer['target'])
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
x = pcad
y = cancer['target']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
svm = SVC()
```

```python
prams = {'C':[0.1,1,10,100,1000],'gamma':[10,1,0.1,0.001,0.0001],'kernel':['rbf']}
grid = GridSearchCV(SVC(),prams)
```

```python
grid.fit(x_train,y_train)
```

```python
pred = grid.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,pred))
print(confusion_matrix(y_test,pred))
```
